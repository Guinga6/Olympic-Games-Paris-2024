```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://olympics.com/en/paris-2024/schedule/grid')
```

```python
html_data = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
html_data
```

```python
html_data = html_data.find_all('div', {'data-testid':'ListRow'})
```

```python
html_data
```

```python
sport_icon = []
sport_name = []
```

```python
for html in html_data:

    icon = html.find('img', {'class': 'emotion-srm-1sv3fyr eph8xjg0'})
    sport_icon.append(icon['src'])

    sport = html.find('span', {'class':'emotion-srm-7id6t6'})
    sport_name.append(sport.text)
```

```python
print(len(sport_icon), len(sport_name))
```

```python
sport_icon.remove(sport_icon[0])
```

```python
sport_name.remove(sport_name[0])
```

```python
df = pd.DataFrame({
        'sport_name' : sport_name,  
        'sport_icon' : sport_icon
        

})
```

```python
df 
```

```python
df = df.replace({ 'Cycling BMX Freestyle': 'BMX Freestyle',
                  'Rhythmic Gymnastics'  : 'Rhyth. Gymnastics',
                    'Trampoline Gymnastics' : 'Trampoline',
                   'Cycling Mountain Bike' : 'Mountain Biking',
                  'Artistic Gymnastics' : 'Gymnastics'
})
```
