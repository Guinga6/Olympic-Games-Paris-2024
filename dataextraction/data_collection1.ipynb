```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.bbc.com/sport/olympics/paris-2024/medals")
```

```python
html_data = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
country_flog = []
```

```python
countryflog = html_data.find_all('div', {'class': 'ssrcss-7kfmgb-BadgeContainer ezmsq4q1'})
```

```python
for flag in countryflog:
    flagx = flag.find('img')
    country_flog.append(flagx['src'])
country_flog
```

```python
names = html_data.find_all('div', {'class' : 'ssrcss-qg0qvg-CellWrapper ef9ipf0'})
names
```

```python
name_of_country = []
```

```python
for name in names:
    namex = name.find('span', {'class': 'ssrcss-pek3um-AbbreviatedCountryName ew4ldjd1'}).text
    name_of_country.append(namex)
```

```python
#name_of_country
```

```python
golden_medal = []
```

```python
goldens = html_data.find_all('div', {'class' : 'ssrcss-z3opd8-CellWrapper ef9ipf0'})
goldens
```

```python
for golden in goldens:
    golden_medal.append(golden.text)
```

```python
#golden_medal
```

```python
silvers = html_data.find_all('div', {'class': 'ssrcss-1yl2exm-CellWrapper ef9ipf0', 'value':'[object Object]'})
```

```python
silvers
```

```python
bronse_medel = []
total_medel = []
silver_medal = []
```
