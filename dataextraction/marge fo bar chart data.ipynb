```python
import pandas as pd 
```

```python
df = pd.read_csv(r"C:\Users\t480\Desktop\price predection project\data_of_medal.csv")
df_medal = df[df['total_medal'] != 0]
df_name_of_country = pd.read_csv(r'C:\Users\t480\Desktop\price predection project\countries.csv')
df = pd.concat([df_name_of_country, df_medal],axis=1)
```

```python
df = df.drop(['name_of_country'], axis=1)
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df
```

```python
list_countries  = df['countries'].to_list()
gold_medal = df['gold_medal'].to_list()
silver_medal = df['silver_medal'].to_list()
bronze_medal = df['bronze_medal'].to_list()
medal_type = ["Gold", 'Silver', 'Bronze']
url_flag = df['url_flag'].to_list()
```

```python
country_list = []
medals = []
type_of_medal = []
flag = []
```

```python
for i in range(len(list_countries)):

    for j in range(3):
        country_list.append(list_countries[i])
        type_of_medal.append(medal_type[j])
        flag.append(url_flag[i])

    medals.append(gold_medal[i])
    medals.append(silver_medal[i])
    medals.append(bronze_medal[i])
        

```

```python
print(len(country_list), len(medals), len(type_of_medal), len(flag))
```

```python
df_1 = pd.DataFrame({
        'countries' : country_list,
        'medal' : medals,
        'medal type':type_of_medal,
        'url_flag' : flag
})
```
